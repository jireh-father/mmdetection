```python
import tensorflow as tf
import os, keras, glob, shutil, cv2
from keras_preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from IPython.display import Image
from pathlib import Path 
from datetime import date

imgSize = 400

model = tf.keras.models.load_model("model_path")

src = "C:/ai/AIFORPET/4.test/blur/blur_2021-04-06/0"
save = "C:/ai/AIFORPET/5.result/hyperemia"



filename = glob.glob(src+'/*.png', recursive = True)

if os.path.exists(save+"/0")==False:
    os.mkdir(save+"/0")

if os.path.exists(save+"/1")==False:
    os.mkdir(save+"/1")

# if os.path.exists(save+"/2")==False:
#     os.mkdir(save+"/2")

# if os.path.exists(save+"/3")==False:
#     os.mkdir(save+"/3")

new_dir_0 =save+'/0/'
new_dir_1 =save+'/1/'
# new_dir_2 =save+'/2/'    
# new_dir_3 =save+'/3/' 

for i in filename:
    img_path = i
    img_name = os.path.basename(i)
    img = tf.keras.preprocessing.image.load_img(i)
    w, h = img.size
    maxsz = max(w, h)
    scale = maxsz / imgSize

    img = img.resize((
        int(w / scale), int(h / scale)
    ))
    imgarray = tf.keras.preprocessing.image.img_to_array(img)
    imgarray = imgarray
    imgarray = np.expand_dims(imgarray, axis=0)
    b = model.predict(imgarray)    
    a = model.predict_classes(imgarray)

    if a[0] == 0:
        print(img_name, b, "0")
        name_0 = new_dir_0+os.path.basename(i)
        shutil.copy(i, name_0)
    elif a[0] == 1:
        print(img_name, b, "1")
        name_1 = new_dir_1+os.path.basename(i)
        shutil.copy(i,name_1)
#     elif a[0] == 2:
#         print(img_name, b, "2")
#         name_2 = new_dir_2+os.path.basename(i)
#         shutil.copy(i, name_2)    
#     elif a[0] == 3:
#         print(img_name, b, "3")
#         name_3 = new_dir_3+os.path.basename(i)
#         shutil.copy(i, name_3)    
```
